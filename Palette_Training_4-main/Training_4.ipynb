```python
df.groupby('State')\
    .mean()\
       .reset_index() 
```

```python
pd.merge(
    df.groupby('State')\
    .mean()\
       .reset_index(),
    gdf,
    on='State'
)
```

```python
gpd.GeoDataFrame(pd.merge(
    df.groupby('State')\
    .mean()\
       .reset_index(),
    gdf,
    on='State'
)).explore(
    column='PricePerIb'
)
```

```python
gpd.GeoDataFrame(pd.merge(
    df.groupby('State')\
    .mean()\
       .reset_index(),
    gdf,
    on='State'
)).explore(
    column='YieldPerCol',
    cmap='rainbow_r'
)
```

```python
df.groupby('State').mean() # 13 years average
```

```python
df.groupby('State')['YieldPerCol'].mean()\
    .sort_values() # 13 years average
```

```python
df.groupby('State')['YieldPerCol'].mean()\
    .sort_values() # 13 years average
```

```python
df.groupby('State')['YieldPerCol'].mean()\
    .sort_values()\
        .plot(kind='bar')
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
df.groupby('State')['YieldPerCol'].mean()\
    .sort_values()\
        .plot(
            kind='barh', 
            figsize=(20,10),
            color='green')
plt.title(
    'Yield by State', 
    size=18,
    color='gray')
plt.xlabel(
    'Yield',
    color='gray',
    size=16)
plt.ylabel(
    'States',
    color='gray',
    size=16)
plt.legend()
plt.show()
```

```python
df_y=df.groupby('State')['YieldPerCol'].mean()\
    .sort_values()\
        .to_frame()\
            .reset_index()
```

```python
df_y
```

```python
sb.barplot(
    x= df_y['YieldPerCol'],
    y = df_y['State'],
    )
plt.title(
    'Yield by State', 
    size=18,
    color='gray')
plt.xlabel(
    'Yield',
    color='gray',
    size=16)
plt.ylabel(
    'States',
    color='gray',
    size=16)

plt.show()
```
