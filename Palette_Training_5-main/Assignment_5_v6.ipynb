```python
# Read in the RM crop yields data
df = pd.read_csv('forecasting_results.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is already loaded with the correct data

crop_types = df['Crop'].unique()
unique_rm_values = df['RM'].unique()
metrics = ['MAE', 'MSE', 'RMSE']
colors = plt.cm.viridis(np.linspace(0, 1, len(unique_rm_values)))

fig, ax = plt.subplots(figsize=(12, 6))

bar_width = 0.25
bar_space = 0.05

for i, rm_value in enumerate(unique_rm_values):
    positions = [j + (bar_width + bar_space) * i for j, _ in enumerate(crop_types)]
    ax.bar(positions, df[df['RM'] == rm_value][metrics[1]], width=bar_width, label=f'RM {rm_value}', color=colors[i])

ax.set_xticks([j + bar_width for j, _ in enumerate(crop_types)])
ax.set_xticklabels(crop_types)
ax.set_ylabel('MAE Values')
ax.set_title('MAE by Crop Type and RM Value')
#ax.legend()

plt.show()

```
